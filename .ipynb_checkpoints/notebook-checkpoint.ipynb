```
import requests, bs4
from bs4 import SoupStrainer
import json
# import textstat
import seaborn as sns
import pandas as pd
import re


loc = 'https://etc.usf.edu/lit2go/books/'

def make_soup(loc):
    res = requests.get(loc)
    res.raise_for_status()
    soup = bs4.BeautifulSoup(res.text, 'html.parser')
    return soup

data = pd.DataFrame(columns = ["Title", "Author", "Level", "Genre"])
soup = make_soup(loc)
links = soup.find_all(class_ = "book_icon", href=True)

for link in links:

    # skip link. it throws error
    if link.get('href') == 'https://etc.usf.edu/lit2go/205/the-story-of-siegfried/':
        continue

    soup = make_soup(str(link.get('href')))
    title = soup.find('h2').get_text().strip()
    author = soup.find('h3').get_text()
    author = re.search('by (.*)', author).group(1)
    # genre = soup.find('strong', content = re.compile('Genre'))
    side_bar = soup.find(id = 'column_secondary')
    # print(side_bar)
    genre = side_bar.find(href = re.compile('genre')).get_text()
    # genre = genre.get_sibling
    # print(genre)
    # genre = re.search('Genre:(.*)', genre).group(1).strip()

    level = soup.find('a', title = re.compile('Flesch')).get_text()

    data.loc[len(data)] = [title, author, level, genre]
```
